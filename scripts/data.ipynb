```python
import os
import h5py
import numpy as np
import matplotlib.pyplot as plt

H5_PATH = "../fashiongen_data/fashiongen_256_256_train.h5"
for i in range(5):


    print("Opening:", os.path.abspath(H5_PATH))
    assert os.path.exists(H5_PATH), "File not found!"

    with h5py.File(H5_PATH, "r") as f:
        # === CORRECT KEYS (based on your file) ===
        images = f["input_image"]                  # (N,256,256,3) uint8
        descriptions = f["input_description"]      # (N,1) bytes
        categories = f["input_category"]           # (N,1) bytes

        print("images:", images.shape, images.dtype)
        print("descriptions:", descriptions.shape, descriptions.dtype)
        print("categories:", categories.shape, categories.dtype)

        # --- read one sample ---
        img = images[i]  # numpy array (256,256,3)

        desc = descriptions[i]
        cat = categories[i]

        # many fields are stored as shape (1,) arrays of bytes -> unwrap + decode
        if isinstance(desc, np.ndarray): desc = desc[0]
        if isinstance(cat, np.ndarray):  cat = cat[0]

        if isinstance(desc, (bytes, bytearray)): desc = desc.decode("utf-8", errors="replace")
        if isinstance(cat, (bytes, bytearray)):  cat  = cat.decode("utf-8", errors="replace")

    print("\nCATEGORY:", cat)
    print("DESCRIPTION:", desc[:300], "..." if len(desc) > 300 else "")

    # --- show image ---
    plt.figure(figsize=(4,4))
    plt.imshow(img)          # img is uint8 RGB
    plt.axis("off")
    plt.title(f"Sample {i} - {cat}")
    plt.show()
```
