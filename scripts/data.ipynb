```python
import h5py

h5_path = "../fashiongen_data/fashiongen_256_256_train.h5"

with h5py.File(h5_path, "r") as f:
    def print_h5(name, obj):
        if isinstance(obj, h5py.Dataset):
            print(name, obj.shape, obj.dtype)
        else:
            print(name)

    f.visititems(print_h5)
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

h5_path = "../fashiongen_data/fashiongen_256_256_train.h5"

with h5py.File(h5_path, "r") as f:
    images = f["images"]
    descriptions = f["descriptions"]
    categories = f["categories"]
    product_names = f["product_names"]

    fig, axes = plt.subplots(1, 5, figsize=(18, 4))

    for i in range(5):
        img = images[i]

        # Convert image to uint8 if needed
        if img.dtype != np.uint8:
            img = (img * 255).astype(np.uint8)

        axes[i].imshow(img)
        axes[i].axis("off")

        # Decode bytes -> str
        desc = descriptions[i].decode("utf-8")
        cat = categories[i].decode("utf-8")
        name = product_names[i].decode("utf-8")

        axes[i].set_title(
            f"{name}\n{cat}",
            fontsize=9
        )

plt.tight_layout()
plt.show()
```
